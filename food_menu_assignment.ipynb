```python
#imports

# user inputs
b = 12.95
c = 10.95
v = 8.95
while True:
    try:
        beef = int(input("Enter number of beef meals: "))
        if beef < 0:
            print("Invalid input!\nOnly positive integers are allowed.")
            continue
        break
    except:
        print("Invalid input!\nOnly integers are allowed.")
        continue
while True:
    try:
        chicken = int(input("Enter number of chicken meals: "))
        if chicken < 0:
            print("Invalid input!\nOnly positive integers are allowed.")
            continue
        break
    except:
        print("Invalid input!\nOnly integers are allowed.")
        continue
while True:
    try:
        vegan = int(input("Enter number of vegan meals: "))
        if vegan < 0:
            print("Invalid input!\nOnly positive integers are allowed.")
            continue
        break
    except:
        print("Invalid input!\nOnly integers are allowed.")
        continue

while True:
    # food gratuity
    g = (0.18 * ((beef * b) + (chicken * c) + (vegan * v)))
    # party number
    n = beef + chicken + vegan

    if n <= 30:
        r = 50
    elif n <= 100:
        r = 100
    elif n <= 150:
        r = 200
    elif n <= 200:
        r = 250
    else:
        print("There is no room large enough to accomodate.")
        print("\n")
        break

    # room tax
    t = r * 0.065 #r is the cost of a room

    # name
    print("\n"*2)
    print("ELIJAH Caterers".center(50,"="))

    # table
    print(f"Number in party             {n}")
    print(f"Room Cost $                               {(r):.2f}")
    print(f"Room Tax $                                {(t):.2f}")
    if beef != 0:
        print(f"Number of beef dinners      {beef}")
        print(f"Cost of beef dinners $                    {(beef * b):.2f}")
    if chicken != 0:
        print(f"Number of chicken dinners   {chicken}")
        print(f"Cost of chicken dinners $                 {(chicken * c):.2f}")
    if vegan != 0:
        print(f"Number of vegan dinners     {vegan}")
        print(f"Cost of vegan dinners $                   {(vegan * v):.2f}")
    print(f"Food Gratuity $                           {(g):.2f}")
    print("-"*50)
    print(f"Total $                                   {(r + t + (beef * b) + (chicken * c) + (vegan * v) + g):.2f}")
    print("\n")
    break
    
print("Process finished with exit code 0")
```
